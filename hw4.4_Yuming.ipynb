```python
# Yuming Gu 4456484 Mechanical Design II homework 4.4
#Problem Statement
#Measure a mechanical spring in surroundings and calcuate the predicted spring rate and maximum shear stress
#Squred-grounded spring mostly used in gel pen, below was a compression springs widely used in gel pen

#A squared-grounded helical compression spring had a outside diameter of 0.24 in (6.1 mm) and wire diameter of 0.026 in (0.66 mm),
#Its free length was 0.75 in (19 mm) and made of stainless steel (G=81 GPa)
#A force of 8.75 N should deflect the spring by 5mm, Maximum workload is 5 lbf (22N), in this problem applied 20 N load

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

#(a)Determine the Spring rate,k
F=float(input("Input the applied force on spring, in N"));                 #F is the applied force on Spring
x=float(input("Input the displacement of the spring, in mm"));             #x is the displacement of the spring
k=F/x;
print('The prediced spring rate is {:.3f} N/mm'.format(k))

#(b)Determine the maximum shear stress
d=float(input("Input the wire diamter of the spring, in mm"));             #d is the spring wire diameter
D=float(input("Input the outside diameter of the spring, in mm"));         #D is the spring outside diameter
C=D/d;                                                                     #C is the spring index
G=float(input("Input the shear modules, in GPa"));                         #G is the shear modules
N_a=(((d*1e-3)**4)*(G*1e+9))/(8*((D*1e-3)**3)*(k*1e+3))                    #N_a is the active turn

Type=float(input("Spring type: plain,no grounded:0; plain,grounded:1; squared,closed/grounded:2"))
if Type == 0:
    N_t=N_a;
elif Type == 1:
    N_t=N_a+1;
else:
    N_t=N_a+2;
                                                                           #N_t is total turn
L_s=N_t*d;                                                                 #L_s is solid length
L_f=float(input("Input the free length of the spring, in mm"));            #L_f is free length
x_max=L_f-L_s;                                                             #x_max is maximum distance
W_max=float(input("Input the maximum workload, in N"));                    #W_max is maximum workload
W_apply=float(input("Input the applied workload, in N"));                  #W_apply is applied workload
F_max=min(W_max,W_apply);                                                  #F_max is maximum force
k_B=(4*C+2)/(4*C-3);                                                       #k_B is Bergtasser Factor
Max_shear=(k_B*8*F_max*(D*1e-3))/((math.pi)*((d*1e-3)**3));                #Max_shear is maximum shear stress
print('The maximum shear stress is {:.3f} MPa'.format(Max_shear*1e-6))


```

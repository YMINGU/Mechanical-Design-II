```python
# Yuming Gu 4456484 Mechanical Design II homework 3.3
#Problem Statement
#Calculate the center-to-center spacing between a pair of gears  
#given the number of teeth on each and their module

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

def Spacing (N_p,N_g,m):
    d1=m*N_p
    d2=m*N_g
    d=d1/2+d2/2
    return d

N_p=float(input("teeth number of the pinion"));# teeth number of the pinion
N_g=float(input("teeth number of the gear"));#teeth number of the gear
m=float(input("gear module"));#gear module
d=Spacing(N_p,N_g,m);
print('The center-to-center spacing is {:.3f} mm'.format(d))
```

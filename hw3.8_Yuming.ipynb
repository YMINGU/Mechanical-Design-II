```python
# Yuming Gu 4456484 Mechanical Design II homework 3.8
#Problem Statement
#Evaluate the face width required for a pair of spur gears 

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

H=float(input("Input Power on the shaft, in KW"));         #P is the input power
n1=float(input("Shaft Rotating Speed, in RPM"));           #n1 is the shaft rotating speed
N1=float(input("Numbers of teeth on pinion"));             #N1 is the numbers of teeth on pinion
N2=float(input("Numbers of teeth on gear"));               #N2 is the numbers of teeth on gear
m=float(input("tooth module, in mm"));                     #m is the tooth module
Fy=float(input("Yield strength of material, in MPa"));     #Y is yield strength of material

#Find the pitch line velocity at the gear mesh, V
n1_rad=n1*(2*math.pi)/60;
r1=((N1*m)/2)*(1e-3);
V=r1*n1_rad;
print('The pitch line velocity at the gear mesh is {:.3f} m/s'.format(V))

#Find the tangential force on the gear, Wt
Tp=H*(1e+3)/n1_rad;                                               #Tp is the torque on the pinion
Wt=Tp/r1;
print('The tangential force on the gear is {:.3f} N'.format(Wt))

#Find the calcuated velocity factor, K_v
Manu_Type=float(input("Enter the types of gear manufacturing methods: hobbed/shaped: 1 cut/milled: 2 and ground:3 "));
if Manu_Type==1:
    k_v=(3.56+(V**0.5))/3.56;
elif Manu_Type==2:
    k_v=(6.1+V)/6.1;
else:
    k_v=((5.56+V**0.5)/5.56)**0.5;
print('The calcuated velocity factor is {:.3f} '.format(k_v))

#Find the shape factor Y
Y=0.1365+(1.273e-2)*N1-(2.185e-4)*(N1**2)+(1.704e-6)*(N1**3)-(4.780e-9)*(N1**4);
print('The shape factor is {:.3f} '.format(Y))

#Find the minimum face width of the two gears in the mesh
F=((k_v*Wt)/(Fy*(1e+6)*m*(1e-3)*Y))*(1e+3);
print('The minimum face width of the two gears in the mesh is {:.3f} mm '.format(F))


    

```

```python
#Yuming Gu 4456484 hw6.4

#Problem Statement
#A cast iron bearing block (below) is going to be used to support a shaft. It will be hung from a steel ceiling joist (e.g., an I-beam).
#The block will support a vertical load of 18 kN and will use four M24 coarse-thread Class 8.8 steel bolts. 
#No washers are used. The dimension "A" of the block is 24 mm and the the ceiling joist flanges are 20 mm thick. 
#Assume a total bolt length of 80 mm and a threaded length of 54 mm. 
#The modulus of elasticity of the cast iron bearing block is 135 GPa.

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

d=24*(1e-3);                                 #d is the bolt diameter, in m
P=18000;                                     #P is the total external load
n=4;                                         #n is number of bolts
S_p=600*(1e+6);                              #S_p is the bolt proof strength for class 8.8 steel bolts
A_t=353*(1e-6);                              #A_t is threaded cross-sectional area for M24 bolt, in m^2
L=80*(1e-3);                                 #L is the total length, in m
w_2=24*(1e-3);                               #w_2 is the thickness of the block, in m
w_1=0;                                       #w_1 is the thickness of the washer, in m
w_3=20*(1e-3);                               #w_3 is the thickness of the ceiling joist flanges, in m
LT=54*(1e-3);                                #LT is the threaded length, in m
E_c=135*(1e+9);                              #E_c is the modulus of elasticity of the cast iron bearing block, in Pa
E_s=200*(1e+9);                              #E_s is 


#Assume permanent joint
Fi=0.90*A_t*S_p*n;                           #Fi is total bolt preload force, in N
print('Total bolt preload force Fi is {:.6f} N'.format(Fi))
print('For each bolt, bolt preload force is {:.6f} N'.format(Fi/n))
A_d=math.pi/4*d**2;
ld=L-LT;
l=w_1+w_2+w_3;
lt=l-ld;
kb=(A_d*A_t*E_s)/(A_d*lt+A_t*ld);
D=1.5*d;
t=-(0.5*(w_1+w_2+w_3)-w_1-w_2);
k_1=(0.5774*math.pi*E_c*d)/(math.log((1.155*(w_1+w_2-t)+D-d)*(D+d)/(1.155*(-t+w_1+w_2)+d+D)/(D-d)));

#Assume cast iron washer
k_2=(0.5774*math.pi*E_c*d)/(math.log((1.155*t+D+2*(w_3-t)*math.tan(30*math.pi/180)-d)*(D+2*(w_3-t)*math.tan(30*math.pi/180)+d)/(1.155*t+d+D+2*(w_3-t)*math.tan(30*math.pi/180))/(D+2*(w_3-t)*math.tan(30*math.pi/180)-d)))
k_3=(0.5774*math.pi*E_s*d)/(math.log((1.155*(w_3+t)+D-d)*(D+d)/(1.155*(w_3+t)+d+D)/(D-d)));
km=1/(1/k_1+1/k_2+1/k_3);
C=kb/(kb+km);                                #C is stiffness Constant
print('Stiffness constant is {:.6f} N'.format(C))

#Lubricated
K=0.18;
T=K*Fi*d/n;                                  #T is wrench torque for each bolt, in N*m
print('Wrench torque for each bolt is {:.6f} N'.format(T))
eta_L=(S_p*A_t-Fi/n)/(C*P/n);                #eta_L is load factor of safety
print('load factor of safety is {:.6f} N'.format(eta_L))
eta_o=Fi/n/((1-C)*P/n);                      #eta_o is factor of safety against joint seperation
print('factor of safety against joint seperation is {:.6f} N'.format(eta_o))
```

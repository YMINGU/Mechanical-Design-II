```python
#Yuming Gu 4456484 Project 3

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

#Section 1: Analysis the Power input on the pedal
mh=70;                                   #mh is the weight of human, in kg
mb=8;                                    #mb is the weight of bike, in kg
g=9.8;                                   #g is the gravitiy acceleration, in N/kg
theta=math.pi/9;                         #theta is degree of the incline
Rr=0.068/2;                              #Rr is the radius of the wheel (27 inches diameter)
Rp=0.17;                                 #Rp is the length of the crank, we choose 170mm
miu=0.014;
Tcf=[36,52];                             #Tcf is the number of teeth of the crankset, we choose 52-36T
Tcr=[11,12,13,14,15,16,17,19,21,23,25]   #Tcr is the number of teeth of the casette, we choose 11-25T
P=300                                    #P is average input power, in W
T=[];
w_p=[];
w_r=[];

#Tension on Chain
for i in Tcf:
    for j in Tcr:
        F=(mh+mb)*g*(math.sin(theta)+miu*math.cos(theta))*(Rr/Rp)*(i/j);
        W_paddle=(P/(F*Rp));
        W_rear=(i/j)*(P/(F*Rp));
        T.append(F);
        w_p.append(W_paddle);
        w_r.append(W_rear);
        
print(T)
print(w_p)
print(w_r)
```

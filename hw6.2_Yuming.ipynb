```python
#Yuming Gu 4456484 hw6.2_Yuming

#Problem statement
#Solve a generic bolted joint problem but with only one material
#The example problem can be two steel plates, both 8 mm thick, no washers,
#held together with an M10 coarse threaded bolt.
#The bolt is 30 mm long and has a threaded length of 26 mm.

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

L=float(input("Total bolt length, in mm"));                            #L is the total bolt length
d=float(input("Bolt diameter, in mm"));                                #d is the bolt diameter
LT=float(input("length of threaded portion of the bolt, in mm"));      #LT is the threaded length
At=float(input("tensile-stress area of threaded portion, in mm^2"));   #At is tensile-stress area of threaded
w1=float(input("thicknesses of washers, in mm (0,1,2)"));              #w1 is washer thickness
w2=float(input("thicknesses of upper plate, in mm"));                  #w2 is upper plate thickness
w3=float(input("thicknesses of lower plate, in mm"));                  #w3 is lower plate thickness
E=207000                                                               #Assume steel, in N/mm^2

#km equation parameters
A=0.78715;
B=0.62873;

Ad=(math.pi)*(d**2)/4;
ld=L-LT;
l=w1+w2+w3
lt=l-ld

kb=(Ad*At*E)/(Ad*lt+At*ld);                                            #kb is bolt stiffness
print('The bolt stiffness is {:.6f} N/micrometer'.format(kb/(1e+3)))

km=E*d*A*math.exp(B*d/l);                                              #km is member stiffness
print('The member stiffness is {:.6f} N/micrometer'.format(km/(1e+3)))

C=kb/(kb+km);                                                          #C is joint stiffness constant
print('The joint stiffness constant is {:.6f} '.format(C))

```

```python
#Yuming Gu 4456484 hw7.4

#Problem Statement
#Write a Python script that assists in the analysis of dual-pad caliper disk brakes.
#These types of brakes are often found on automobiles and bicycles.
#Estimate the stopping torque applied by a set of bicycle brakes. 

#Basic Variable Setting
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math

#In this problem, the input variables are specific geometry, coefficient of friction, and actuating force.
#The output variables are torque applied by a set of disk brakes.

#Ref:https://bikerumor.com/aasq-92-braking-force-vs-rotor-size-lever-leverage-more-hydraulic-brake-answers/

#Example 16-3
F=3748;                            #F is the actuating force, in Lbf
theta1=36*math.pi/180;             #theta 1 is in rad
theta2=144*math.pi/180;            #theta 2 is in rad
ro=5.5;                            #ro is in inch
ri=3.875;                          #ri is in inch
f=0.37;

P_a=F/(theta2-theta1)/ri/(ro-ri);
T=P_a*(theta2-theta1)*f*ri*(ro**2-ri**2)/2;
print('The torque applied by disk brakes is {:.6f} N'.format(T))

#To estimate the stopping torque applied by a set of bicycle brakes
F2=3400;                            #F is the actuating force, in Lbf
theta3=60*math.pi/180;             #theta 1 is in rad
theta4=120*math.pi/180;            #theta 2 is in rad
ro2=150;                            #ro is in mm
ri2=130;                          #ri is in mm
f2=0.37;

P_a2=F2/(theta4-theta3)/ri2/(ro2-ri2);
T2=P_a2*(theta4-theta3)*f2*ri2*(ro2**2-ri2**2)/2;
print('The stopping torque applied by a set of bicycle brakes is {:.6f} N'.format(T2/1e+3))
```
